```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bangalore_data = pd.read_csv('bangaloredata.csv', encoding='ascii')

# Drop the duplicate rows based on 'last_updated' keeping the first occurrence
bangalore_data_dropped = bangalore_data.drop_duplicates(subset=['last_updated'], keep='first')

# Convert the 'last_updated' column to datetime
bangalore_data_dropped['last_updated'] = pd.to_datetime(bangalore_data_dropped['last_updated'])

# Set the 'last_updated' column as the index
bangalore_data_dropped.set_index('last_updated', inplace=True)
```

```python
# Plotting 'feels like' temperature, actual temperature, and humidity over time
plt.figure(figsize=(14, 7))

# Actual Temperature
sns.lineplot(data=bangalore_data_dropped, x=bangalore_data_dropped.index, y='temperature_celsius', label='Actual Temperature (Celsius)', color='blue')

# Feels Like Temperature
sns.lineplot(data=bangalore_data_dropped, x=bangalore_data_dropped.index, y='feels_like_celsius', label='Feels Like Temperature (Celsius)', color='red')

# Humidity
sns.lineplot(data=bangalore_data_dropped, x=bangalore_data_dropped.index, y='humidity', label='Humidity (%)', color='green', linestyle='--')
plt.title('Actual vs Feels Like Temperature and Humidity over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (Celsius) / Humidity (%)')
plt.legend()
plt.tight_layout()
plt.show()
```
